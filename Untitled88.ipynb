### 1. Load the data

Let's start by loading the data. Assuming your data is in a CSV file, you can use the following code to load it into a pandas DataFrame. Remember to replace `'your_data.csv'` with the actual path to your file.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('/content/dataset.csv')
```

### 2. Understand the data

Let's get a first look at the data by displaying the first few rows and checking the data types and non-null values.

```python
# Display the first 5 rows
print("First 5 rows of the DataFrame:")
display(df.head())

# Get information about the DataFrame
print("\nDataFrame Info:")
df.info()
```

```python
df.tail()
```

### 3. Clean the data

Let's start by checking for missing values.

```python
# Check for missing values
print("\nMissing values in each column:")
display(df.isnull().sum())
```

```python
# Drop rows with missing values in 'Credit_Score'
df.dropna(subset=['Credit_Score'], inplace=True)

print("Missing values after dropping rows with missing 'Credit_Score':")
display(df.isnull().sum())
```

Let's also check for duplicate rows.

```python
# Check for duplicate rows
print("\nNumber of duplicate rows:")
display(df.duplicated().sum())
```

### 4. Analyze the data

Let's start by looking at descriptive statistics for the numerical columns.

```python
# Display descriptive statistics for numerical columns
print("\nDescriptive Statistics for Numerical Columns:")
display(df.describe())
```

Let's also explore the categorical variables.

```python
# Display unique values and their counts for categorical columns
print("\nUnique values and counts for categorical columns:")
for col in df.select_dtypes(include='object').columns:
    print(f"\nColumn: {col}")
    display(df[col].value_counts())
```

Let's visualize the distribution of some key variables.

```python
# Set a style for the plots
sns.set_style('whitegrid')

# 1. Distribution of Credit_Score
plt.figure(figsize=(8, 5))
sns.countplot(x='Credit_Score', data=df, palette='viridis')
plt.title('Distribution of Credit Score')
plt.xlabel('Credit Score')
plt.ylabel('Count')
plt.show()

# 2. Distribution of Age
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# 3. Distribution of Annual_Income
plt.figure(figsize=(10, 6))
sns.histplot(df['Annual_Income'], bins=30, kde=True, color='salmon')
plt.title('Distribution of Annual Income')
plt.xlabel('Annual Income')
plt.ylabel('Frequency')
plt.show()

# 4. Distribution of Credit_Mix
plt.figure(figsize=(8, 5))
sns.countplot(x='Credit_Mix', data=df, palette='viridis')
plt.title('Distribution of Credit Mix')
plt.xlabel('Credit Mix')
plt.ylabel('Count')
plt.show()

# 5. Distribution of Interest_Rate
plt.figure(figsize=(10, 6))
sns.histplot(df['Interest_Rate'], bins=30, kde=True, color='lightgreen')
plt.title('Distribution of Interest Rate')
plt.xlabel('Interest Rate')
plt.ylabel('Frequency')
plt.show()
```

Let's visualize the relationships between some pairs of variables.

```python
# 1. Credit_Score vs Annual_Income
plt.figure(figsize=(10, 6))
sns.boxplot(x='Credit_Score', y='Annual_Income', data=df, palette='viridis')
plt.title('Credit Score vs Annual Income')
plt.xlabel('Credit Score')
plt.ylabel('Annual Income')
plt.show()

# 2. Credit_Score vs Interest_Rate
plt.figure(figsize=(10, 6))
sns.boxplot(x='Credit_Score', y='Interest_Rate', data=df, palette='viridis')
plt.title('Credit Score vs Interest Rate')
plt.xlabel('Credit Score')
plt.ylabel('Interest Rate')
plt.show()

# 3. Credit_Score vs Credit_Utilization_Ratio
plt.figure(figsize=(10, 6))
sns.boxplot(x='Credit_Score', y='Credit_Utilization_Ratio', data=df, palette='viridis')
plt.title('Credit Score vs Credit Utilization Ratio')
plt.xlabel('Credit Score')
plt.ylabel('Credit Utilization Ratio')
plt.show()

# 4. Credit_Score vs Credit_History_Age
plt.figure(figsize=(10, 6))
sns.boxplot(x='Credit_Score', y='Credit_History_Age', data=df, palette='viridis')
plt.title('Credit Score vs Credit History Age')
plt.xlabel('Credit Score')
plt.ylabel('Credit History Age')
plt.show()

# 5. Credit_Score vs Num_of_Delayed_Payment
plt.figure(figsize=(10, 6))
sns.boxplot(x='Credit_Score', y='Num_of_Delayed_Payment', data=df, palette='viridis')
plt.title('Credit Score vs Number of Delayed Payments')
plt.xlabel('Credit Score')
plt.ylabel('Number of Delayed Payments')
plt.show()
```

Let's visualize the relationships between multiple variables.

```python
# 1. Relationship between Annual_Income, Interest_Rate, and Credit_Score
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Annual_Income', y='Interest_Rate', hue='Credit_Score', data=df, palette='viridis', alpha=0.6)
plt.title('Annual Income vs Interest Rate colored by Credit Score')
plt.xlabel('Annual Income')
plt.ylabel('Interest Rate')
plt.show()

# 2. Relationship between Credit_Utilization_Ratio, Num_of_Delayed_Payment, and Credit_Score
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Credit_Utilization_Ratio', y='Num_of_Delayed_Payment', hue='Credit_Score', data=df, palette='viridis', alpha=0.6)
plt.title('Credit Utilization Ratio vs Number of Delayed Payments colored by Credit Score')
plt.xlabel('Credit Utilization Ratio')
plt.ylabel('Number of Delayed Payments')
plt.show()

# 3. Relationship between Age, Credit_History_Age, and Credit_Score
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Age', y='Credit_History_Age', hue='Credit_Score', data=df, palette='viridis', alpha=0.6)
plt.title('Age vs Credit History Age colored by Credit Score')
plt.xlabel('Age')
plt.ylabel('Credit History Age')
plt.show()

# 4. Relationship between Annual_Income, Credit_History_Age, and Credit_Score
plt.figure(figsize=(12, 8))
sns.boxplot(x='Credit_Score', y='Annual_Income', hue='Credit_Mix', data=df, palette='viridis')
plt.title('Credit Score vs Annual Income by Credit Mix')
plt.xlabel('Credit Score')
plt.ylabel('Annual Income')
plt.show()


# 5. Pairplot of selected numerical variables colored by Credit_Score
selected_cols = ['Annual_Income', 'Interest_Rate', 'Credit_Utilization_Ratio', 'Num_of_Delayed_Payment', 'Credit_Score']
sns.pairplot(df[selected_cols], hue='Credit_Score', palette='viridis')
plt.suptitle('Pairwise Relationships of Selected Variables by Credit Score', y=1.02)
plt.show()
```

```python
# 6. Pie chart of Credit_Score distribution
credit_score_counts = df['Credit_Score'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(credit_score_counts, labels=credit_score_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('viridis', len(credit_score_counts)))
plt.title('Distribution of Credit Score')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# 7. Distribution of Monthly_Inhand_Salary
plt.figure(figsize=(10, 6))
sns.histplot(df['Monthly_Inhand_Salary'], bins=30, kde=True, color='purple')
plt.title('Distribution of Monthly Inhand Salary')
plt.xlabel('Monthly Inhand Salary')
plt.ylabel('Frequency')
plt.show()

# 8. Count of Payment_of_Min_Amount
plt.figure(figsize=(8, 5))
sns.countplot(x='Payment_of_Min_Amount', data=df, palette='viridis')
plt.title('Count of Payment of Minimum Amount')
plt.xlabel('Payment of Minimum Amount')
plt.ylabel('Count')
plt.show()

# 9. Relationship between Num_of_Loan and Credit_Score
plt.figure(figsize=(10, 6))
sns.boxplot(x='Credit_Score', y='Num_of_Loan', data=df, palette='viridis')
plt.title('Credit Score vs Number of Loans')
plt.xlabel('Credit Score')
plt.ylabel('Number of Loans')
plt.show()

# 10. Relationship between Outstanding_Debt and Credit_Score
plt.figure(figsize=(10, 6))
sns.boxplot(x='Credit_Score', y='Outstanding_Debt', data=df, palette='viridis')
plt.title('Credit Score vs Outstanding Debt')
plt.xlabel('Credit Score')
plt.ylabel('Outstanding Debt')
plt.show()

# 11. Count of Occupation
plt.figure(figsize=(12, 6))
sns.countplot(y='Occupation', data=df, order=df['Occupation'].value_counts().index, palette='viridis')
plt.title('Count of Occupation')
plt.xlabel('Count')
plt.ylabel('Occupation')
plt.show()
```

### 5. Summarize Findings

Based on the exploratory data analysis, here are some key findings:

*   **Data Loading and Cleaning:**
    *   The dataset contains information on credit behavior with 30752 entries and 28 columns.
    *   Initially, there was one missing value in the 'Credit_Score' column, which was successfully handled by dropping the corresponding row.
    *   There were no duplicate rows found in the dataset.

*   **Descriptive Statistics and Distributions:**
    *   The 'Credit_Score' distribution shows that 'Standard' credit scores are the most frequent, followed by 'Poor' and then 'Good'.
    *   The 'Age' distribution appears to be somewhat normally distributed with a peak around the early 30s.
    *   'Annual_Income' is right-skewed, indicating that most individuals have lower annual incomes, with fewer individuals having very high incomes.
    *   The 'Credit_Mix' is predominantly 'Standard', with a good number of 'Good' and 'Bad' credit mixes as well.
    *   'Interest_Rate' shows a varied distribution with some peaks at lower rates and a spread across higher rates.
    *   'Monthly_Inhand_Salary' is also right-skewed, similar to 'Annual_Income'.
    *   The 'Payment_of_Min_Amount' is split between 'Yes' and 'No', with a smaller category of 'NM' (Not Mentioned).
    *   The 'Occupation' distribution shows a relatively even spread across different professions, with 'Teacher', 'Architect', and 'Lawyer' being among the most frequent.

*   **Relationships between Variables:**
    *   **Credit Score vs. Annual Income:** Individuals with 'Good' credit scores tend to have higher annual incomes compared to those with 'Standard' or 'Poor' credit scores, although there is significant overlap.
    *   **Credit Score vs. Interest Rate:** Individuals with 'Good' credit scores generally have lower interest rates compared to those with 'Standard' or 'Poor' credit scores.
    *   **Credit Score vs. Credit Utilization Ratio:** 'Good' credit scores are associated with lower credit utilization ratios, while 'Poor' credit scores tend to have higher ratios.
    *   **Credit Score vs. Credit History Age:** Individuals with 'Good' credit scores generally have a longer credit history age.
    *   **Credit Score vs. Number of Delayed Payments:** 'Good' credit scores are associated with fewer delayed payments, while 'Poor' credit scores have a higher number of delayed payments.
    *   **Credit Score vs. Outstanding Debt:** 'Good' credit scores appear to be associated with lower outstanding debt compared to 'Standard' and 'Poor' credit scores.
    *   **Credit Score vs. Credit Mix:** The boxplot of 'Credit Score' vs 'Annual Income' by 'Credit Mix' shows that within each credit mix category, higher annual income is generally associated with better credit scores.

These findings provide a good starting point for understanding the factors that may influence credit scores in this dataset. Further analysis and potentially modeling could be done based on these insights.

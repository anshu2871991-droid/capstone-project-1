```python
missing_percentage = (df.isnull().sum() / len(df)) * 100
missing_df = pd.DataFrame({'missing count': df.isnull().sum(), 'missing %': missing_percentage})
missing_df = missing_df.sort_values('missing %', ascending=False)
print(missing_df)
```

```python
plt.figure(figsize=(10,8))
sns.histplot(df['Age'],kde=True,color='purple',bins=20)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.show()
```

```python
gender_counts = df['Gender'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Gender Distribution')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Primary_Device', data=df, order = df['Primary_Device'].value_counts().index)
plt.title('Distribution of Primary Devices')
plt.xlabel('Primary Device')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot( y='Avg_Daily_Screen_Time_hr', data=df)
plt.title('Distribution of Average Daily Screen Time by Primary Device')
plt.xlabel('Primary Device')
plt.ylabel('Average Daily Screen Time (hours)')
plt.xticks(rotation=45)
plt.show()
```

# ***bivarate***

```python
gender_device = df.groupby(['Gender', 'Primary_Device']).size().unstack()
gender_device.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Primary Device Usage by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Age', y='Avg_Daily_Screen_Time_hr', data=df)
plt.title('Age vs. Average Daily Screen Time')
plt.xlabel('Age')
plt.ylabel('Average Daily Screen Time (hours)')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Health_Impacts', y='Avg_Daily_Screen_Time_hr', data=df)
plt.title('Screen Time Distribution by Health Impacts')
plt.xlabel('Health Impacts')
plt.ylabel('Average Daily Screen Time (hours)')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='Age', y='Avg_Daily_Screen_Time_hr', data=df)
plt.title('Average Screen Time by Age')
plt.xlabel('Age')
plt.ylabel('Average Daily Screen Time (hours)')
plt.grid(True)
plt.show()
```

```python
sns.pairplot(df, hue='Gender')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
numerical_columns = df.select_dtypes(include=['number']).columns
plt.figure(figsize=(20, 8))
for i, column in enumerate(numerical_columns):
    plt.subplot(2, 3, i + 1)
    sns.histplot(df[column], kde=True)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.tight_layout()
plt.show()
```

```python
Categorical_columns = df.select_dtypes(include=['object']).columns
plt.figure(figsize=(20, 8))
for i, column in enumerate(Categorical_columns):
    plt.subplot(2, 3, i + 1)
    sns.countplot(x=column, data=df)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.tight_layout()
plt.show()
```

## ***1. Know Your Data***

### Import Libraries

````python
# Import Python Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ignore warnings
import warnings
warnings.filterwarnings('ignore')
````

### Dataset Loading

````python
# Load the Play Store Data
df_playstore = pd.read_csv('Play Store Data.csv')

# Load the User Reviews Data
df_reviews = pd.read_csv('User Reviews.csv')
````

### Dataset First Look

````python
# Display the first 5 rows of the Play Store Data
print("Play Store Data:")
display(df_playstore.head())

# Display the first 5 rows of the User Reviews Data
print("\nUser Reviews Data:")
display(df_reviews.head())
````

### Dataset Info

````python
# Get information about the Play Store Data
print("Play Store Data Info:")
display(df_playstore.info())

# Get information about the User Reviews Data
print("\nUser Reviews Data Info:")
display(df_reviews.info())
````

### Duplicate Values

````python
# Count duplicate rows in Play Store Data
playstore_duplicates = df_playstore.duplicated().sum()
print(f"Number of duplicate rows in Play Store Data: {playstore_duplicates}")

# Count duplicate rows in User Reviews Data
reviews_duplicates = df_reviews.duplicated().sum()
print(f"Number of duplicate rows in User Reviews Data: {reviews_duplicates}")
````

### Missing Values/Null Values

````python
# Check for missing values in Play Store Data
print("Missing values in Play Store Data:")
display(df_playstore.isnull().sum())

# Check for missing values in User Reviews Data
print("\nMissing values in User Reviews Data:")
display(df_reviews.isnull().sum())
````

### Merge Datasets

````python
# Merge the two dataframes on the 'App' column
df_merged = pd.merge(df_playstore, df_reviews, on='App')

# Display the first few rows of the merged dataframe
print("Merged Data:")
display(df_merged.head())

# Get information about the merged dataframe
print("\nMerged Data Info:")
display(df_merged.info())

# Check for missing values in the merged dataframe
print("\nMissing values in Merged Data:")
display(df_merged.isnull().sum())
````

### Handle Missing Values and Data Cleaning

Let's first address the missing values in the merged dataframe. We can see from the `isnull().sum()` output that 'Translated_Review', 'Sentiment', 'Sentiment_Polarity', 'Sentiment_Subjectivity', and 'Rating' have missing values. For the sentiment-related columns and 'Translated_Review', we can fill the missing values with a placeholder like 'Unknown' or a relevant value if appropriate. For 'Rating', we could consider imputation strategies, but given the small number of missing values relative to the dataset size, we might also consider dropping them or imputing with the mean or median. Let's start by filling the missing sentiment-related values and 'Translated_Review' with 'Unknown'. For 'Rating', let's impute with the median.

````python
# Fill missing values in sentiment-related columns and Translated_Review with 'Unknown'
df_merged[['Translated_Review', 'Sentiment', 'Sentiment_Polarity', 'Sentiment_Subjectivity']] = df_merged[['Translated_Review', 'Sentiment', 'Sentiment_Polarity', 'Sentiment_Subjectivity']].fillna('Unknown')

# Impute missing 'Rating' values with the median
median_rating = df_merged['Rating'].median()
df_merged['Rating'] = df_merged['Rating'].fillna(median_rating)

# Check missing values again to confirm
print("Missing values after handling:")
display(df_merged.isnull().sum())
````

Now, let's clean and transform the 'Size', 'Installs', and 'Price' columns to numerical formats.

````python
# Clean 'Size' column
df_merged['Size'] = df_merged['Size'].astype(str).str.replace('M', 'e6').str.replace('k', 'e3').replace('Varies with device', np.nan)
df_merged['Size'] = pd.to_numeric(df_merged['Size'])

# Clean 'Installs' column
df_merged['Installs'] = df_merged['Installs'].str.replace(',', '').str.replace('+', '').astype(int)

# Clean 'Price' column
df_merged['Price'] = df_merged['Price'].str.replace('$', '').astype(float)

# Display data types to confirm transformations
print("\nData types after cleaning:")
display(df_merged.info())
````
